# Gemini API Simple Project

This notebook demonstrates how to use the latest `google-genai` SDK to interact with Gemini models using an API key.

```python
import os
from dotenv import load_dotenv
from google import genai
```

```python
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")
model_id = os.getenv("GOOGLE_MODEL", "gemini-2.0-flash-lite")

if api_key:
    print(f"API Key loaded. Target Model: {model_id}")
else:
    print("Error: GOOGLE_API_KEY not found!")
```

```python
with open("prompt.md", "r", encoding="utf-8") as f:
    prompt_text = f.read()

print("--- Prompt Content ---")
print(prompt_text)
```

```python
prompt_text = "대한민국이 언제 1인당 소득 4만달러 돌파하나요?"
```

```python
client = genai.Client(api_key=api_key)

print(f"Calling Gemini ({model_id})...")
response = client.models.generate_content(
    model=model_id,
    contents=prompt_text
)

print("\n--- Gemini Response ---")
print(response.text)
```
